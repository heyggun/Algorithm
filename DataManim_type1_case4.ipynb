Q. 흡연자와 비흡연자 각각 charges의 상위 10% 그룹의 평균의 차이는?

```python
smoker_high = df.loc[df['smoker']=='yes']['charges'].quantile(0.9)
nonsmoker_high = df.loc[df['smoker']=='no']['charges'].quantile(0.9)

print(smoker_high, nonsmoker_high)

smoker_mean = df.loc[(df['smoker']=='yes') & (df['charges']>= smoker_high)]['charges'].mean()
nonsmoker_mean = df.loc[(df['smoker']=='no') & (df['charges']>= nonsmoker_high)]['charges'].mean()


Ans = smoker_mean - nonsmoker_mean
Ans
```

Q. bedrooms의 빈도가 가장 높은 값을 가지는 데이터들의 price의 상위 10%와하위 10% 값의 차이를 구하여라

```python
fre_bed = df['bedrooms'].value_counts().index[0]

high_price = df.loc[df['bedrooms']==fre_bed]['price'].quantile(0.9)
low_price = df.loc[df['bedrooms']==fre_bed]['price'].quantile(0.1)

Ans = high_price - low_price
Ans
```

## 레드 와인 퀄리티 예측 데이터

```python
DataUrl = 'https://raw.githubusercontent.com/Datamanim/datarepo/main/redwine/train.csv'
df = pd.read_csv(DataUrl)
display(df.head(3))
```

Q. quality 값이 3인 그룹과 8인 데이터그룹의 각 컬럼별 독립변수의 표준편차 값의 차이를 구할때 그값이 가장 큰 컬럼명을 구하여라

```python
qual_3 = df.loc[df['quality']==3].std()
qual_8 = df.loc[df['quality']==8].std()


temp = pd.concat([qual_3, qual_8], axis=1).rename(columns={0:'qual_3', 1:'qual_8'})
temp['gap'] = temp['qual_3'] - temp['qual_8']
temp['abs_gap'] = temp['gap'].apply(lambda x: abs(x))
temp.sort_values(by='abs_gap', ascending=False).index[0]
```

```python
answer = (df.loc[df.quality ==8].std() -df.loc[df.quality ==3].std()).sort_values().index[-1]
print(answer)
```
